```python
# Dtypes 
df["date"] = pd.to_datetime(df["date"])

# World Cup 2018 end date
df = df.loc[(df['date'] > '2018-07-15')]
df = df.sort_values(by='date')
df
```

```python
# Select Teams 
teams = ['Qatar', 'England', 'Argentina', 'France', 'Spain', 'Belgium', 'Brazil', 'Portugal',
          'Ecuador', 'Iran', 'Saudi Arabia', 'Australia', 'Costa Rica', 'Canada', 'Serbia', 'Ghana',
          'Senegal', 'United States', 'Mexico', 'Denmark', 'Germany', 'Morocco', 'Switzerland', 'Uruguay',
          'Netherlands', 'Wales', 'Poland', 'Tunisia', 'Japan', 'Croatia', 'Cameroon', 'South Korea']
```

```python
wc_22_home = df[df.home_team.isin(teams)]
wc_22_away = df[df.away_team.isin(teams)]
```

### Home Record

```python
wc_22_home.head()
```

```python
# Results
WDL = [wc_22_home['home_score'] > wc_22_home['away_score'],
       wc_22_home['home_score'] == wc_22_home['away_score'],
       wc_22_home['home_score'] < wc_22_home['away_score']]
result = ['W', 'D', 'L']

wc_22_home['result'] = np.select(WDL, result) 
wc_22_home.head()
```

```python
# games played 
games_played = wc_22_home.groupby(['home_team']).size()
games_played = games_played.to_frame().reset_index()
games_played.columns = ['home_team', 'games_played']
games_played.head()

# wdl tally
tally = wc_22_home.groupby(['home_team'])['result'].value_counts().unstack().fillna(0).reset_index()
tally = tally[['home_team', 'W', 'D', 'L']]

# merge gp and tally
game_tally = pd.merge(games_played, tally, on="home_team", how='right')
game_tally.head()
```

```python
# new dataframe
home_rec = wc_22_home.groupby(['home_team'])['home_score', 'away_score'].sum().reset_index()
home_rec = home_rec.rename(columns={"home_score": "goals_for",
                                    "away_score": "goals_against"})
# merge 
home_rec = pd.merge(game_tally, home_rec, on="home_team")

home_rec.head()
```

### AWAY Record

```python
wc_22_away.head()
```

```python
# Results
WDL2 = [wc_22_away['home_score'] < wc_22_away['away_score'],
       wc_22_away['home_score'] == wc_22_away['away_score'],
       wc_22_away['home_score'] > wc_22_away['away_score']]
result2 = ['W', 'D', 'L']

wc_22_away['result'] = np.select(WDL2, result2) 
wc_22_away.head()
```

```python
# games played 
games_played2 = wc_22_away.groupby(['away_team']).size()
games_played2 = games_played2.to_frame().reset_index()
games_played2.columns = ['away_team', 'games_played']

# wdl tally
tally2 = wc_22_away.groupby(['away_team'])['result'].value_counts().unstack().fillna(0).reset_index()
tally2 = tally2[['away_team', 'W', 'D', 'L']]

# merge gp and tally
game_tally2 = pd.merge(games_played2, tally2, on="away_team", how='right')
game_tally2.head()
```

```python
# Convert W,D,L columns to int
game_tally2[['W', 'D', 'L']] = game_tally2[['W', 'D', 'L']].astype(int)

# new dataframe
away_rec = wc_22_away.groupby(['away_team'])['away_score', 'home_score'].sum().reset_index()
away_rec = away_rec.rename(columns={"away_score": "goals_for",
                                    "home_score": "goals_against"})
# merge 
away_rec = pd.merge(game_tally2, away_rec, on="away_team")

away_rec.head()
```

# Stats into the 2022 Fifa World Cup

```python
home_rec = home_rec.rename(columns={"home_team": "team"})
away_rec = away_rec.rename(columns={"away_team": "team"})
```

```python
display('home df', home_rec.head(3))
display('away df', away_rec.head(3))
```

```python
# Combine Dataframe
wc_df = pd.concat([home_rec, away_rec]).groupby(['team']).sum().reset_index()
```

```python
# Assigning teams to groups

dict = {'Qatar': 'A', 'Netherlands': 'A', 'Senegal': 'A', 'Ecuador': 'A',
        'England': 'B', 'United States': 'B', 'Iran': 'B', 'Wales': 'B',
        'Argentina': 'C', 'Saudi Arabia': 'C', 'Mexico': 'C', 'Poland': 'C',
        'France': 'D', 'Australia': 'D', 'Denmark': 'D', 'Tunisia': 'D',
        'Spain': 'E', 'Germany': 'E', 'Japan': 'E', 'Costa Rica': 'E',
        'Belgium': 'F', 'Croatia': 'F', 'Canada': 'F', 'Morocco': 'F',
        'Brazil': 'G', 'Serbia': 'G', 'Switzerland': 'G', 'Cameroon': 'G',
        'Portugal': 'H', 'Ghana': 'H', 'Uruguay': 'H', 'South Korea': 'H'}

wc_df['group'] = wc_df['team'].apply(lambda x : dict[x])
wc_df = wc_df.reindex(columns=['group']+wc_df.columns[:-1].tolist()) # move to first
```

```python
# Assigning teams to confederations

AFC = ['Qatar', 'Iran', 'Saudi Arabia', 'Australia', 'Japan', 'South Korea']
       
UEFA = ['Netherlands', 'England', 'Wales', 'Poland', 'France', 'Denmark', 'Spain',
        'Germany', 'Belgium', 'Croatia', 'Serbia', 'Switzerland', 'Portugal']
        
CAF = ['Senegal', 'Tunisia', 'Morocco', 'Cameroon', 'Ghana'] 
        
CONMEBOL = ['Ecuador', 'Argentina', 'Costa Rica', 'Brazil', 'Uruguay']    
        
CONCACAF = ['United States', 'Mexico', 'Canada'] 
    
d = {'AFC': AFC, 'UEFA': UEFA, 'CAF': CAF,
     'CONMEBOL': CONMEBOL, 'CONCACAF': CONCACAF}
    
d1 = {k: oldk for oldk, oldv in d.items() for k in oldv}
wc_df['confederation'] = wc_df['team'].map(d1) 


# shift column 'Name' to first position
shift_column = wc_df.pop('confederation')
# insert column using insert(position,column_name,
# shift_column) function
wc_df.insert(2, 'confederation', shift_column)
```

### Equations for Stats

```python
### Stats bound ###

# goal diff
wc_df['goal_diff'] = wc_df['goals_for'] - wc_df['goals_against']

# Scoring ratio
wc_df['goal_scoring_ratio'] = (wc_df['goals_for'] / wc_df['goals_against']).round(2)

# Goals per Game
wc_df['g_pg'] = (wc_df['goals_for'] / wc_df['games_played']).round(2)
wc_df['ga_pg']= (wc_df['goals_against'] / wc_df['games_played']).round(2)

# Goal Diff per game
wc_df['gdiff_pg'] = (wc_df['g_pg'] - wc_df['ga_pg']).round(2)

# EST Win % 
est_win_percentage = (wc_df['goal_scoring_ratio']**2 / 
                     (wc_df['goal_scoring_ratio']**2 + 1)) - (wc_df['D']/wc_df['games_played'])
wc_df['est_win_%'] = est_win_percentage.round(2)

# true_win_%
wc_df['true_win_%'] = (wc_df['W']/wc_df['games_played']).round(2)

# Absoulte error 
wc_df['absolute_error'] = (wc_df['true_win_%'] - wc_df['est_win_%']).round(3)
```

```python
#Mean AD (MAD)

from sklearn.metrics import mean_absolute_error as mae

actual = wc_df['true_win_%']
calculated = wc_df['est_win_%']

error = mae(actual, calculated).round(3)
print("Mean absolute error : " + str(error))
```

```python
MAD_win = (wc_df['goal_scoring_ratio']**6 /
                 (wc_df['goal_scoring_ratio']**6 + 1)) - (wc_df['D']/wc_df['games_played'])
wc_df['win_%'] = MAD_win.round(2)
```
